### Определение выбросов

```python
def outlier(column):
    iqr = column.quantile(0.75) - column.quantile(0.25)
    return column[(column < column.quantile(0.25) - 1.5*iqr)|(column>column.quantile(0.75) + 1.5*iqr)]
```

```python
for col in df:
    print(outlier(df[col]))
    print()
```

Выбросы пристсуют в колонках: возраст (9), температура (экстремально низкие и высокие), пульс (экстремально высокий), respiratory_rate (высокий)
