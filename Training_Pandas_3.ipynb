### Задание 1

```python
log = pd.read_csv("visit_log.csv", sep=';')
```

```python
log.traffic_source.value_counts()
```

```python
log.head()
```

```python
def source_type(row):
    if row.traffic_source in {'yandex', 'google'}:
        return 'organic'
    elif row.traffic_source in {'paid', 'email'}:
        if row.region == 'Russia':
            return 'ad'
        else:
            return 'other'
    return row.traffic_source
```

```python
log['source_type'] = log.apply(source_type, axis=1)
```

### Задание 3

```python
ratings = pd.read_csv("ml-latest-small/ratings.csv")
```

```python
ratings.head()
```

```python
groupped_ratings = ratings.groupby("userId").agg({'timestamp':['count', 'min', 'max']}).rename(columns = {'count':'logs', 'min': 'mintime', 'max': 'maxtime' })
```

```python
groupped_ratings.columns = groupped_ratings.columns.droplevel(0)
```

```python
groupped_ratings.head()
```

```python
groupped_ratings['livetime']=groupped_ratings.apply(lambda row: row.maxtime - row.mintime, axis = 1)
```

```python
groupped_ratings[groupped_ratings.logs >= 100].livetime.mean()
```
